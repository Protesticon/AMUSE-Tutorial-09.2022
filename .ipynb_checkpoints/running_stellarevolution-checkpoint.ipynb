How to run a stellar evolution code
====================

Here we will generate initial conditions for a stellar evolution code. We will adopt a parametrized stellar evolution code.

For reference you can read chapter 3 of Portegies Zwart & McMillan 2018 (2018araa.book.....P). 

With this tutorial you will learn
- how to generate inital conditions (mass-function)
- how to run a parametrized stellar evolution code
- channels and data transfer
- plotting results

```python
%matplotlib inline
import numpy
from matplotlib import pyplot
from amuse.units import units
```

We start by setting up a simulation by specifying the initial mass function. What would you say is the biggest difference between these IMFs?

```python
from amuse.datamodel import Particles
n_stars = 1024
mmin = 0.1 | units.MSun
mmax = 100 | units.MSun

from amuse.lab import new_kroupa_mass_distribution
mkroupa = new_kroupa_mass_distribution(n_stars,
                                           mass_min=mmin, 
                                           mass_max=mmax)
k_stars = Particles(mass=mkroupa)
from amuse.lab import new_salpeter_mass_distribution
msalpeter = new_salpeter_mass_distribution(n_stars, 
                                           mass_min=mmin, 
                                           mass_max=mmax)
s_stars = Particles(mass=msalpeter)
print("mean-mass for Kroupa=", numpy.mean(k_stars.mass))
print("mean-mass for Salpeter=", numpy.mean(s_stars.mass))

pyplot.hist(numpy.log10(k_stars.mass.value_in(units.MSun)))
pyplot.title('Kroupa IMF')
pyplot.yscale('log')
pyplot.xlim(-1, 2)
pyplot.figure()
pyplot.hist(numpy.log10(s_stars.mass.value_in(units.MSun)))
pyplot.title('Salpeter IMF')
pyplot.yscale('log')
pyplot.xlim(-1, 2)
pyplot.show()
```

Now that the mass functions are determined, we can evolve the stars in time. For this we start two stellar evolution codes, both parametrized, and evolve the stars in time.

```python
from amuse.community.seba.interface import SeBa
def start_stellar_code(stars):
    stellar = SeBa()
    stellar.particles.add_particles(stars)
    channels = {"to_stars": stellar.particles.new_channel_to(stars), 
                "to_stellar": stars.new_channel_to(stellar.particles)}
    return stellar, channels
kstellar, kchannels = start_stellar_code(k_stars)
sstellar, schannels = start_stellar_code(s_stars)
```

Now the stellar evolution code are declared. Now evolve in time and calculate the mean mass as a function of time.

```python
times = 10**numpy.arange(0.0, 4.0, 0.1) | units.Myr
mmean = []
for time in times:
    kstellar.evolve_model(time)
    kchannels["to_stars"].copy()
    sstellar.evolve_model(time)
    schannels["to_stars"].copy()
    mmean.append(numpy.mean(k_stars.mass)/numpy.mean(s_stars.mass))
kstellar.stop()
sstellar.stop()
print("mean-mass for Kroupa=", numpy.mean(k_stars.mass))
print("mean-mass for Salpeter=", numpy.mean(s_stars.mass))
```

```python
from amuse.plot import plot
plot(times, mmean)
pyplot.ylabel("relative mean mass")
pyplot.semilogx()
pyplot.show()
```

```python
from amuse.plot import scatter
scatter(s_stars.temperature, s_stars.luminosity, c='r')
scatter(k_stars.temperature, k_stars.luminosity, c='b', s=3)
pyplot.xlim(2.e+4, 2000)
pyplot.ylim(1.e-5, 1000)
pyplot.loglog()
pyplot.show()
```

You have calculated the evolution of a stellar population with solar metalicity.

### Assignment 2

```python
def start_stellar_code(stars, z=None):
    stellar = SeBa()
    if z:
        stellar.parameters.metallicity = z
    stellar.particles.add_particles(stars)
    channels = {"to_stars": stellar.particles.new_channel_to(stars), 
                "to_stellar": stars.new_channel_to(stellar.particles)}
    return stellar, channels

mkroupa = new_kroupa_mass_distribution(n_stars,
                                       mass_min=mmin, 
                                       mass_max=mmax)

k_stars_z002  = Particles(mass=mkroupa)
k_stars_z0002 = Particles(mass=mkroupa)


kstellar_z002,  kchannels_z002  = start_stellar_code(k_stars_z002, z=0.02)
kstellar_z0002, kchannels_z0002 = start_stellar_code(k_stars_z0002, z=0.002)
```

```python
times = 10**numpy.arange(0.0, 4.0, 0.1) | units.Myr
mass_z002 = []
mass_z0002 = []
for time in times:
    kstellar_z002.evolve_model(time)
    kchannels_z002["to_stars"].copy()
    kstellar_z0002.evolve_model(time)
    kchannels_z0002["to_stars"].copy()
    mass_z002.append(numpy.mean(k_stars_z002.mass.value_in(units.MSun)))
    mass_z0002.append(numpy.mean(k_stars_z0002.mass.value_in(units.MSun)))
kstellar_z002.stop()
kstellar_z0002.stop()
print("mean-mass for Kroupa z0.02 =", numpy.mean(k_stars_z002.mass))
print("mean-mass for Kroupa z0.002 =", numpy.mean(k_stars_z0002.mass))
```

```python
plot(times, mass_z002, label='z=0.02')
plot(times, mass_z0002, label='z=0.002')
pyplot.ylabel("Mean Mass of Stellar Population")
pyplot.semilogx()
pyplot.legend()
pyplot.show()
```

The high metallicity population usually have smaller mass. The high metallicity stars have less fuels and live shorter.

### Question 1

```python
compact_types = [10,11,12,13,14]

mask = numpy.zeros(len(k_stars.stellar_type)).astype('bool')
for compact_type in compact_types:
    mask = mask | (k_stars.stellar_type == (compact_type|units.stellar_type))
print('Kroupa IMF:', mask.sum(), 'of the 1024 stars are compact objects.')

mask = numpy.zeros(len(s_stars.stellar_type)).astype('bool')
for compact_type in compact_types:
    mask = mask | (s_stars.stellar_type == (compact_type|units.stellar_type))
print('Salpeter IMF:', mask.sum(), 'of the 1024 stars are compact objects.')

mask = numpy.zeros(len(k_stars_z002.stellar_type)).astype('bool')
for compact_type in compact_types:
    mask = mask | (k_stars_z002.stellar_type == (compact_type|units.stellar_type))
print('Metallicity=0.02:', mask.sum(), 'of the 1024 stars are compact objects.')

mask = numpy.zeros(len(k_stars_z0002.stellar_type)).astype('bool')
for compact_type in compact_types:
    mask = mask | (k_stars_z0002.stellar_type == (compact_type|units.stellar_type))
print('Metallicity=0.002:', mask.sum(), 'of the 1024 stars are compact objects.')
```

Different IMF have statistically different number of compact objects, but the metallicity does not.

```python
compact_types = [10,11,12,13,14]


mask = k_stars.stellar_type == (1|units.stellar_type)
print('Kroupa IMF:', mask.sum(), 'of the 1024 stars are main sequence stars.')


mask = s_stars.stellar_type == (1|units.stellar_type)
print('Salpeter IMF:', mask.sum(), 'of the 1024 stars are main sequence stars.')


mask = k_stars_z002.stellar_type == (1|units.stellar_type)
print('Metallicity=0.02:', mask.sum(), 'of the 1024 stars are main sequence stars.')


mask = k_stars_z0002.stellar_type == (1|units.stellar_type)
print('Metallicity=0.002:', mask.sum(), 'of the 1024 stars are main sequence stars.')
```

These numbers are not statistically different. The fraction are all close to 95%.
